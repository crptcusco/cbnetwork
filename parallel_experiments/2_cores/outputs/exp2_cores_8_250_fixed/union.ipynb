```python
import pandas as pd

# Cargar los archivos CSV
df104 = pd.read_csv("data104.csv")
df190 = pd.read_csv("data190.csv")

# Obtener los identificadores únicos de sample de cada archivo
unique_samples_104 = sorted(df104['i_sample'].unique())
unique_samples_190 = sorted(df190['i_sample'].unique())

# Seleccionar los 100 samples del primer archivo y 150 del segundo
selected_samples_104 = unique_samples_104[:100]
selected_samples_190 = unique_samples_190[:150]

# Filtrar las filas que corresponden a los samples seleccionados
df104_subset = df104[df104['i_sample'].isin(selected_samples_104)].copy()
df190_subset = df190[df190['i_sample'].isin(selected_samples_190)].copy()

# Reasignar los números de sample de forma consecutiva
# Para el primer conjunto, asignamos de 1 a 100
mapping_104 = {old: new for new, old in enumerate(selected_samples_104, start=1)}
df104_subset['i_sample'] = df104_subset['i_sample'].map(mapping_104)

# Para el segundo, asignamos de 101 a 250
mapping_190 = {old: new for new, old in enumerate(selected_samples_190, start=101)}
df190_subset['i_sample'] = df190_subset['i_sample'].map(mapping_190)

# Combinar ambos DataFrames
df_combined = pd.concat([df104_subset, df190_subset], ignore_index=True)

# Guardar el DataFrame combinado en un nuevo archivo CSV
df_combined.to_csv("data.csv", index=False)
```
