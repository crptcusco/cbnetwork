```python
import matplotlib.pyplot as plt

# Datos
cores = [2, 4, 8, 12, 16, 20, 24]
serial = [81.5565, 81.0568, 81.3557, 81.9810, 81.3595, 83.1686, 81.9260]
parallel_a = [41.2481, 20.2501, 11.3147, 8.5077, 7.6244, 7.3980, 7.1845]
parallel_b = [41.3109, 20.3644, 11.0604, 8.5285, 7.5250, 7.4625, 7.3355]

# Crear figura y ejes
plt.figure(figsize=(10, 6))

# Graficar las líneas
plt.plot(cores, serial, marker='o', linestyle='-', label='Serial', color='black')
plt.plot(cores, parallel_a, marker='s', linestyle='--', label='Parallel A', color='blue')
plt.plot(cores, parallel_b, marker='^', linestyle='-.', label='Parallel B', color='red')

# Configurar etiquetas y título
plt.xlabel("Number of Cores")
plt.ylabel("Execution Time (s)")
plt.title("Execution Time vs Number of Cores")
plt.legend()
plt.grid(True)

# Mostrar gráfico
plt.show()
```
