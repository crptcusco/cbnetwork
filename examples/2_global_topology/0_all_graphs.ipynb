# All Allowed Topologies by GlobalTopology

```python
from cbnetwork.globaltopology import GlobalTopology

GlobalTopology.show_allowed_topologies()
```

```python
import matplotlib.pyplot as plt

def safe_plot_topology(obj):
    if obj is None:
        fig, ax = plt.subplots()
        ax.text(0.5, 0.5, 'No topology (None)', ha='center', va='center')
        ax.axis('off')
        return
    try:
        obj.plot_topology()
    except Exception as e:
        fig, ax = plt.subplots()
        ax.text(0.5, 0.5, f'Plot failed: {e}', ha='center', va='center')
        ax.axis('off')
```

## Complete Directed Graph

```python
N_NODES = 6
N_EDGES = None
V_TOPOLOGY = 1

# omit n_edges when it's None to avoid passing an explicit None that some generators don't accept
o_global_topology = GlobalTopology.generate_sample_topology(v_topology=V_TOPOLOGY, n_nodes=N_NODES)
safe_plot_topology(o_global_topology)
type(o_global_topology)
```

```python
N_NODES = 6
V_TOPOLOGY = 2

o_global_topology = GlobalTopology.generate_sample_topology(v_topology=V_TOPOLOGY, n_nodes=N_NODES)
safe_plot_topology(o_global_topology)
type(o_global_topology)
```
